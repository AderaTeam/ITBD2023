```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./DATA/Hack/train_dataset_train.csv', sep=";")
data.head()
```

```python
theme_group_counts = data['Группа тем'].value_counts()
theme_group_counts
```

```python
plt.bar(theme_group_counts.index, theme_group_counts)
plt.xticks(rotation = 90)
plt.hlines(theme_group_counts.median(), xmin=0, xmax=26, label='med', colors='r')
plt.hlines(theme_group_counts.mean(), xmin=0, xmax=26, label='mean', colors='orange')
plt.legend()
plt.show()
```
